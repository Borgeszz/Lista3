```python
#Felipe Borges - 2111351 - Lista 3
#Github - https://github.com/Borgeszz/Lista3.git

import numpy as np
import pandas as pd

from google.colab import drive
drive.mount('/content/drive')

breast_cancer = pd.read_csv('/content/drive/MyDrive/class_breast_cancer.csv')

breast_cancer = pd.DataFrame(breast_cancer)

display(breast_cancer)
```

```python
#Inspecione as colunas do conjunto de dados (Nome das colunas e tipo de dados)

breast_cancer.info()
```

```python
#Escolha uma coluna com valores numéricos e apague todas as linhas que possuam valor nulo

breast_cancer['concavity'] = breast_cancer['concavity'].replace(0,np.nan)

breast_cancer['concavity'] = breast_cancer['concavity'].dropna()
```

```python
#Calcule a média desta mesma coluna

med = breast_cancer['concavity'].mean()
print(med)
```

```python
#Escolha um critério para eliminar as linhas que possuam valores muito distantes da média e crie um novo Dataframe.

crit1 = breast_cancer['concavity'] >=0.01
crit2 = breast_cancer['concavity'] <=0.3

breast_cancer = breast_cancer[crit1]
breast_cancer = breast_cancer[crit2]
```

```python
#Crie um novo Dataframe que contenha apenas 5 colunas do conjunto anterior.

breast_cancer2 = breast_cancer.drop(['radius','texture','perimeter','area','smoothness'],axis=1)
```

```python
#Escolha duas colunas numéricas e crie uma nova coluna que siga a regra: (Coluna A - Coluna B)*número aleatório (entre 0 e 1).

breast_cancer2['coluna_nova'] = breast_cancer2['compacity']-breast_cancer2['fractal']
breast_cancer2['coluna_nova'] = breast_cancer2['coluna_nova']*0.4
```

```python
#Embaralhe aleatoriamente o dataframe (índices e valores).

breast_cancer_random = breast_cancer2.sample(frac=1)
```

```python
#Ordene o Dataframe seguindo um critério definido por você (Explique o critério)

breast_cancer2 = breast_cancer2[['fractal','compacity','symmetry','coluna_nova','target']] #Organizei de maneira aleatoria, sem nenhum criterio especifico
breast_cancer2.sort_values(by = ['symmetry']).reset_index(drop=True) #Organizei a coluna da simetria em ordem crescente, depois resetei o index para ficar de acordo com a nova ordem
```

```python
#Escolha uma coluna numérica e um coluna com strings e utilize a função Groupby para mostrar os totais da coluna.

junçao = breast_cancer2.groupby(['compacity','target']).count()
```

```python
#Utilize a função MAPS para fazer algum tipo de substituição no Dataframe (explique)

breast_cancer2['target'] = breast_cancer2['target'].map({'M':'Metastasis','B':'Luminal B'}) #Usei a funcao MAPS e troquei o nome dos estagios do cancer de mama
```
